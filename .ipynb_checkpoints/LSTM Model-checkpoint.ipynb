```python
from yahoo_fin.stock_info import get_data
```

```python
apple = get_data("aapl", start_date="01/01/2023", end_date="30/09/2023", index_as_date = True, interval="1d")
apple.head()
```

```python
print(apple)
```

```python
apple = apple.reset_index()
print(apple)
```

```python
apple = apple.drop(['index','adjclose','ticker'],axis = 1)
apple.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(apple.close)
```

```python
ma100 = apple.close.rolling(10).mean()
ma100
```

```python
plt.figure(figsize = (12,6))
plt.plot(apple.close)
plt.plot(ma100,'r')
```

```python
ma200 = apple.close.rolling(20).mean()
```

```python
plt.figure(figsize = (12,6))
plt.plot(apple.close)
plt.plot(ma100,'r')
plt.plot(ma200,'g')
```

```python
#Splitting Data into Training and Testing
import pandas as pd
print(apple.shape)

dtrain = pd.DataFrame(apple['close'][0:int(len(apple)*0.70)])
dtest = pd.DataFrame(apple['close'][int(len(apple)*0.70):int(len(apple))])

print(dtrain.shape)
print(dtest.shape)
```
